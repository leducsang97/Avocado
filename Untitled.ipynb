```python
data = pd.read_csv('avocado.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.head()
```

```python
df = data
df = df.rename(index=str, columns={"4046": "Medium", "4225": "Large", "4770":"Extra Large"})
newCol = df['Total Volume'] - df["Medium"]-df["Large"]-df["Extra Large"]
df.insert(loc=7, column="Other", value=newCol)
df.head()
```

```python
df = df.drop(['Total Bags', 'Small Bags', 'Large Bags','XLarge Bags'], axis=1)
```

```python
df_Cali = df[df['region']=='California']
df_Cali.head()
```

```python
df_Cali = df_Cali.drop(['region'], axis=1)
```

```python
df_Cali.head()
```

```python
Q1 = (df_Cali['Date'].dt.quarter==1)
Q2 = (df_Cali['Date'].dt.quarter==2)
Q3 = (df_Cali['Date'].dt.quarter==3)
Q4 = (df_Cali['Date'].dt.quarter==4)
a_2018 = (df_Cali['year']==2018)
a_2017 = (df_Cali['year']==2017)
a_2016 = (df_Cali['year']==2016)
a_2015 = (df_Cali['year']==2015)

Organic = (df_Cali['type']=='organic')
Conv = (df_Cali['type']=='conventional')
```

```python
df_Cali.insert(loc=2, column="Q4", value=Q4)
df_Cali.insert(loc=2, column="Q3", value=Q3)
df_Cali.insert(loc=2, column="Q2", value=Q2)
df_Cali.insert(loc=2, column="Q1", value=Q1)

df_Cali.insert(loc=12, column="2015", value=a_2015)
df_Cali.insert(loc=12, column="2016", value=a_2016)
df_Cali.insert(loc=12, column="2017", value=a_2017)
df_Cali.insert(loc=12, column="2018", value=a_2018)
```

```python
df_Cali.insert(loc=14, column="organic", value=Organic)
df_Cali.insert(loc=14, column="conventional", value=Conv)
```

```python
time_2017 = pd.Timestamp(year=2017, month=1, day=1)
df_Cali_train = df_Cali[df_Cali['Date']<time_2017]
df_Cali_predict = df_Cali[df_Cali['Date']>=time_2017]

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import train_test_split
#from sklearn import cross_validation
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
import warnings
warnings.filterwarnings('ignore')
```

```python
x = df_Cali_train.drop(['Date','type', 'Total Volume','Unnamed: 0','year'], axis = 1)
y = df_Cali_train.AveragePrice
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
mean_squared_error(y_test, reg.predict(x_test))
```

```python
reg.score(x_test, y_test)
```

```python
from sklearn import linear_model
lasso = linear_model.Lasso(alpha=0.1).fit(x_train, y_train)
lasso.predict(x_test)
mean_squared_error(y_test, lasso.predict(x_test))
```

```python
lasso.score(x_test, y_test)
```
